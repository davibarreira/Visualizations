# Tutorial - Learning Julia

```julia
using DataFrames, RDatasets
```

```julia
iris = RDatasets.dataset("datasets", "iris")
```

```julia
first(iris,5)
```

```julia
col = "SepalLength"
iris[!,Symbol(col)]
col  = names(iris)
iris[!, col[1]]
```

```julia
iris[:,[:SepalLength]]
iris[:,r"SepalLength"]
describe(iris[:,[:SepalLength]])
```

```julia
using Plots, StatsPlots
plotly()
```

```julia
p1 = Plots.histogram(iris[:SepalLength])
p2 = Plots.boxplot(iris[!,:SepalLength])

plot(p1,p2,layout=(1,2))
```
