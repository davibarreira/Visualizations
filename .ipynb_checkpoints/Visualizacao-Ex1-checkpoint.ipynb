# Visualização de Dados - Exercício 1
Realizar um scatter plot utilizando uma biblioteca similar ao ggplot e outra
com Grammar of Graphics como o Vega-lite.

```julia
using Pkg
Pkg.add("Plots")
# Utilizando VegaLite
Pkg.add("VegaLite")
Pkg.add("VegaDatasets")
Pkg.build("NodeJS")
# Biblioteca similar ao ggplot
Pkg.add("Gadfly")
Pkg.add("RDatasets")
Pkg.add("PyCall")
```

```julia
using Gadfly, RDatasets
iris = dataset("datasets", "iris")
```

```julia
p = plot(iris, x=:SepalLength, y=:SepalWidth, color=:Species, Geom.point);
p
```

```julia
using VegaLite
```

```julia

iris |>
@vlplot(
    :point,
    x=:SepalLength,
    y=:SepalWidth,
    color=:Species,
    width=400,
    height=400
)
    
```
